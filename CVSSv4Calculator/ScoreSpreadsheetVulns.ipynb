```python
import pandas as pd
```

```python
ss_vulns = pd.read_csv('spreadsheet_vulns.csv')
vects = ss_vulns['Vector'].values.tolist()
```

```python
SHEET_ID = '1V9NRbUdTKgx4iTRthJsjf3Pp59JqavH-44nwHEMmDO8'
SHEET_NAME = 'ToBeScored'
url = f'https://docs.google.com/spreadsheets/d/{SHEET_ID}/gviz/tq?tqx=out:csv&sheet={SHEET_NAME}'
df = pd.read_csv(url)
```

```python
vects = df['Vector'].values.tolist()
vects = [v.replace("E:E", "E:A") for v in vects]
```

```python
import selenium
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.webdriver.chrome.options import Options

options = Options()
options.headless = True
options.add_argument("--window-size=1920,1200")
```

```python
DRIVER_PATH = '/Users/bedwards/Downloads/chromedriver_mac64/chromedriver'
driver = webdriver.Chrome(executable_path=DRIVER_PATH, options=options)
driver.get("https://assuremoss.github.io/CVSSv4Calculator/index.html")
preset_select = Select(driver.find_element(By.ID, "preset_select"))
```

```python
rows = []
for i,vect in enumerate(vects):
    row = {}
    driver.get("https://assuremoss.github.io/CVSSv4Calculator/index.html#" + vect)
    preset_select = Select(driver.find_element(By.ID, "preset_select"))
    for opt in preset_select.options:
        opt_val = opt.get_attribute("value")
        preset_select.select_by_value(opt_val)
        opt_score = float(driver.find_element(By.ID, "final_score").text.split('/')[0].strip())
        row[opt_val] = opt_score
    row["vect"] =  vect
    rows.append(row)
    if i % 10==0:
        print(i, "/", len(vects))
```

```python
pd.DataFrame(rows).to_csv("./calc_spreadsheet_vulns.csv")
```
